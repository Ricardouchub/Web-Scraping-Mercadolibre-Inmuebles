```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# --- Configuración de Selenium ---
service = Service()
options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
driver = webdriver.Chrome(service=service, options=options)

# Abrimos una página inicial para configurar la ubicación
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# --- Configuración de Selenium ---
service = Service()
options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
driver = webdriver.Chrome(service=service, options=options)

# Abrimos una página inicial para configurar la ubicación
driver.get("https://listado.mercadolibre.cl/inmuebles/departamentos/venta/")

# --- Pausa para Acción Manual ---
input("""
------------------------------------------------------------------
ACCIÓN MANUAL REQUERIDA:
1. La ventana de Chrome está abierta.
2. Si aparece un pop-up, agrega una ubicación manualmente.
3. Una vez que la página se vea normal, vuelve a esta terminal y 
   PRESIONA ENTER para continuar.
------------------------------------------------------------------
""")

print("Continuando con el scraping automatizado...")
propiedades_bruto = []

# --- CORRECCIÓN 1: Bucle ajustado para 42 páginas ---
for pagina in range(1, 43): 
    
    # --- CORRECCIÓN 2: Lógica para manejar la URL de la página 1 ---
    if pagina == 1:
        url_completa = "https://listado.mercadolibre.cl/inmuebles/departamentos/venta/rm-metropolitana/santiago/_NoIndex_True"
    else:
        numero_url = (pagina - 1) * 48 + 1
        url_completa = f"https://listado.mercadolibre.cl/inmuebles/departamentos/venta/rm-metropolitana/santiago/_Desde_{numero_url}_NoIndex_True"
    
    print(f"\nScrapeando página {pagina} de 42: {url_completa}")
    driver.get(url_completa)
    
    try:
        # Esperamos a que los anuncios estén visibles
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "ui-search-layout__item"))
        )
    except Exception as e:
        print(f"No se cargaron los resultados en la página {pagina}. Saltando...")
        continue

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    resultados = soup.find_all('li', class_='ui-search-layout__item')

    # --- Lógica de Extracción Bruta ---
    for item in resultados:
        texto_completo = item.get_text(separator=' | ', strip=True)
        if texto_completo:
            propiedades_bruto.append({'datos_brutos': texto_completo})
            
    # Pausa de cortesía para no sobrecargar el servidor
    time.sleep(2)

driver.quit()

import os # Importante: añade esta línea al principio de tu script

# ... (todo tu código de scraping anterior va aquí) ...

# --- Reporte Final ---
print("\nScraping finalizado. Creando el archivo CSV...")

if propiedades_bruto:
    df = pd.DataFrame(propiedades_bruto)

    # Definir la carpeta y el nombre del archivo
    carpeta_salida = 'data'
    nombre_archivo = 'datos_raw_inmuebles_santiago.csv'
    ruta_completa = os.path.join(carpeta_salida, nombre_archivo)

    # Asegurarse de que la carpeta de salida exista
    os.makedirs(carpeta_salida, exist_ok=True)

    # Guardar el DataFrame en la ruta especificada
    df.to_csv(ruta_completa, index=False, encoding='utf-8-sig')

    print(f"¡ÉXITO! Archivo '{ruta_completa}' creado con {len(df)} registros.")
else:
    print("No se extrajo ninguna propiedad. El DataFrame está vacío.")
```
